```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
#loading Customer sales data CSV
sales_df = pd.read_csv('/content/drive/MyDrive/Pandas_intro/company_sales_data.csv')
sales_df
```

```python
#Comparing facecream and toothpaste sales data using a bar chart.
x = np.arange(len(sales_df['month_number']))
month_number = ['jan', 'feb', 'mar', 'april', 'may', 'june', 'july', 'aug', 
         'sept','oct','nov','dec']
y1 = sales_df['toothpaste']
plt.bar(x -0.2, y1, width=0.4,)

y2 = sales_df['facecream']
plt.bar(x +0.2,y2, width=0.4, )

plt.title('FACECREAM AND TOOTHPASTE SALES')
plt.xticks(x, month_number)
plt.xlabel('months')
plt.ylabel('amount')
plt.legend(sales_df, loc= 'upper left')
```

```python
#Assuming the average monthly sales of bathing soap is 10k,
#visualizing the sales data of bathing soap of all months, 
#showing how they differ from the average sale.

x_position=np.arange(len(sales_df['month_number']))
month_number= ['jan','feb', 'march', 'april', 'may', 'june', 'july','aug','sept', 'oct', 'nov', 'dec']       
x_error = 1
y_error = 1
y = sales_df['bathingsoap']
plt.bar(x,y,)
plt.errorbar(x,y, xerr= x_error, yerr= y_error, capsize=10)
plt.xticks(x,month_number)
plt.title('MONTHLY SALES OF BATHINGSOAP USING AVERAGE')
plt.xlabel('MONTHS')
plt.ylabel('SALES OF BATHINGSOAP')
plt.show()
```

```python
#Calculate the total sales data for the year 
#of  each product and represent it on a pie chart
sum_total= [sales_df['facecream'].sum(),sales_df['facewash'].sum(),sales_df['toothpaste'].sum(),
            sales_df['bathingsoap'].sum(),sales_df['shampoo'].sum(),sales_df['moisturizer'].sum()]
items = ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
plt.pie(sum_total,autopct= '%d%%')
plt.title('TOTAL OF EACH ITEM SOLD')
plt.legend(items,loc='upper right')
plt.axis('equal')
plt.show()
```
