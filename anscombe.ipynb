```python
#importing dependences
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
```

```python
#READING ANSCOMBE CSV
anscombe_df = pd.read_csv('/content/gdrive/MyDrive/Pandas_intro/anscombe_quartet_data.csv')
anscombe_df
```

```python
#Descriptive statistics of anscombe_df
anscombe_df.describe()

```

```python
#Correlation data of anscombe_df
anscombe_df.corr()
```

```python
#creating subplots showing how each x column correlates with each y column.
#using the 4different methods
#FIRST APPROACH TO SUBPLOTS
fig = plt.figure(figsize=(8,6))
x = [0,14]
y = [0,10.84]
x1 = anscombe_df['x1']
y1 = anscombe_df['y1']
plt.subplot(2,2,1)
plt.title('(FIRST X1,Y1)')
plt.xlabel('x1')
plt.ylabel('y1')
plt.scatter(x1,y1, color='red')
plt.plot(x,y,color= 'blue')

x = [0,14]
y = [0,9.26]
x2 =anscombe_df['x2']
y2 =anscombe_df['y2']
plt.subplot(2,2,2)
plt.title('FIRST X2, Y2')
plt.xlabel('x2')
plt.ylabel('y2')
plt.scatter(x2,y2, color= 'red')
plt.plot(x,y,color='blue')

x = [0,14]
y = [0,12.75]
x3=anscombe_df['x3']
y3=anscombe_df['y3']
plt.subplot(2,2,3)
plt.title('FIRST X3, Y3')
plt.xlabel('x3')
plt.ylabel('y3')
plt.scatter(x3,y3,color='red')
plt.plot(x,y,color='blue')

x = [0,14]
y = [0,12.50]
x4=anscombe_df['x4']
y4=anscombe_df['y4']
plt.subplot(2,2,4)
plt.title('FIRST, X4, Y4')
plt.xlabel('x4')
plt.ylabel('y4')
plt.scatter(x4,y4,color='red')
plt.plot(x,y,color='blue')
#adjusting
plt.subplots_adjust(wspace=0.5, hspace=0.5)
```

```python
#SECOND APPROACH TO SUBPLOTS
fig,((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(8,6))
x = [0,14]
y = [0,10.84]
ax1.plot(x,y,color='black')
ax1.scatter(x1, y1, color='blue')
ax1.set_title('SECOND X1,Y1')
ax1.set_xlabel('X1')
ax1.set_ylabel('Y1')

x = [0,14]
y = [0,9.26]
ax2.plot(x,y,color='black')
ax2.scatter(x2,y2,color='blue')
ax2.set_title('SECOND X2,Y2')
ax2.set_xlabel('X2')
ax2.set_ylabel('Y2')

x = [0,14]
y = [0,12.75]
ax3.plot(x,y,color='black')
ax3.scatter(x3,y3, color='blue')
ax3.set_title('SECOND X3,Y3')
ax3.set_xlabel('X3')
ax3.set_ylabel('Y3')

x = [0,14]
y = [0,12.50]
ax4.plot(x,y,color='black')
ax4.scatter(x4,y4,color='blue')
ax4.set_title('SECOND X4,Y4')
ax4.set_xlabel('X4')
ax4.set_ylabel('Y4')
plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)

```

```python
#THIRD APPROACH TO SUBPLOTS
fig= plt.figure(figsize=(8, 6))
x = [0,14]
y = [0,10.48]
ax5 = fig.add_subplot(221, title='THIRD X1, Y1', xlabel='X1', ylabel= 'Y1')
ax5.plot(x, y, color='green')
ax5.scatter(x1,y1, color= 'yellow')

x = [0,14]
y = [0, 9.26]
ax6 = fig.add_subplot(222, title='THIRD X2, Y2', xlabel= 'X2', ylabel= 'Y2')
ax6.plot(x, y, color='green')
ax6.scatter(x2,y2, color= 'yellow')

x = [0,14]
y = [0,12.75]
ax7 = fig.add_subplot(223, title='THIRD X3, Y3', xlabel= 'X3', ylabel= 'Y3')
ax7.plot(x, y, color='green')
ax7.scatter(x3,y3, color= 'yellow')

x = [0,14]
y = [0,12.50]
ax8 = fig.add_subplot(224, title='THIRD X4, Y4', xlabel= 'X4', ylabel= 'Y4')
ax8.plot(x, y, color='green')
ax8.scatter(x4,y4, color= 'yellow')
plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)


```
