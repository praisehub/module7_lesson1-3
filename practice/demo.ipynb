```python
#after installing matplotlib,import for use
import matplotlib
from matplotlib import pyplot as plt
```

```python
#ploting a simple barchart
x_values = [1,2,3,4,5]
y_values = [2,4,6,8,10]

plt.plot (x_values, y_values)
plt.show()
```

```python
#plot the folling points on a line graph (1,2), (4,6), (3,2), (5,1)
import numpy as np
x_values = np.array([1, 4, 3, 5])
y_values = ([2, 6, 2, 1])
plt.plot(x_values,y_values)
plt.show()
```

```python
#plotting multiple lines
days = [0,1,2,3,4,5,6]
bees_allowance = [20, 30, 40, 50, 60, 70, 80]
panuels_allowance = [10, 15, 20, 25, 35, 45, 55]
plt.plot(days,bees_allowance, linestyle = 'dotted', color = 'black', marker = 'o')
plt.plot(panuels_allowance, linestyle = 'dashed', color = 'red', marker = '*')
#naming the axes
plt.xlabel('Days of the week')
plt.ylabel('Allowance in dollars')
#creating a plot title
plt.title('DAILY ALLOWANCE')
plt.show()
```
