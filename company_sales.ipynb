```python
#IMPORTING MY DEPENDENCES
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#READIND COMPANY_SALES CSV
company_df = pd.read_csv('/content/gdrive/MyDrive/Pandas_intro/company_sales_data.csv')
company_df
```

```python
#Comparing all the product sales data for every month using the multiple line plot. 
fig = plt.figure(figsize=(10,8))
x = company_df['month_number']
y1 = company_df['facecream']
plt.plot(x,y1,color='purple', linestyle=':', marker='o')

x = company_df['month_number']
y2 = company_df['bathingsoap']
plt.plot(x,y2,color='blue', linestyle= '--', marker= '*')

x = company_df['month_number']
y3 = company_df['facewash']
plt.plot(x,y3,color='black',linestyle= '-', marker= '+')

x = company_df['month_number']
y4 = company_df['toothpaste']
plt.plot(x,y4,color= 'red', linestyle= ':', marker= 'o')

x = company_df['month_number']
y5 = company_df['moisturizer']
plt.plot(x,y5,color= 'violet', linestyle= ':', marker= '*')

x= company_df['month_number']
y6 = company_df['shampoo']
plt.plot(x,y6,color= 'green', linestyle= '--', marker= '+')

#LEGEND AT THE TOP LEFT CORNER OF THE AXES SHEET
plt.legend(['facecream','bathingsoap', 'facewash','toothpaste','moisturizer','shampoo'])

#LEBELLING MY AXES
plt.title('BODY CARE PRODUCTS MONTHLY SALES')
plt.xlabel('MONTHS')
plt.ylabel('SALES OF BODY CARE PRODUCTS')
plt.show

```
